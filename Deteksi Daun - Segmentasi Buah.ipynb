```python
# Fathur Rasyid Irimawa 202431036
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fathur Rasyid Irimawa 202431036
img = cv2.imread('JerukNipis.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Fathur Rasyid Irimawa
lower_jeruk = np.array([35, 50, 50])
upper_jeruk = np.array([85, 255, 255])
mask_jeruk = cv2.inRange(img_hsv, lower_jeruk, upper_jeruk)
segmentasi_jeruk = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_jeruk)
```

```python
# Fathur Rasyid Irimawa
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(img)
ax[0].set_title('Citra Asli')

ax[1].imshow(segmentasi_jeruk)
ax[1].set_title('Segmentasi Jeruk Nipis')

plt.show()
```
