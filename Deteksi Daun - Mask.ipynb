```python
# Fathur Rasyid Irimawa 202431036
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fathur Rasyid Irimawa 202431036
img = cv2.imread('JerukNipis.jpeg')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Fathur Rasyid Irimawa 202431036
lower_green = np.array([30, 40, 40])
upper_green = np.array([90, 255, 255])
mask_jeruk = cv2.inRange(img_hsv, lower_green, upper_green)
```

```python
# Fathur rasyid Irimawa 202431036
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(img)
ax[0].set_title('Citra Asli')

ax[1].imshow(mask_jeruk, cmap='gray')
ax[1].set_title('Mask Jeruk Nipis')

plt.show()
```
