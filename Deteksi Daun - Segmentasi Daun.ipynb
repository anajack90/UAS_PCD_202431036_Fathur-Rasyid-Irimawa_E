```python
# Fathur Rasyid Irimawa 202431036
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fathur Rasyid Irimawa
img = cv2.imread('JerukNipis.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Fathur Rasyid Irimawa 202431036
lower_hijau = np.array([25, 30, 30])
upper_hijau = np.array([90, 255, 255])
mask_awal = cv2.inRange(img_hsv, lower_hijau, upper_hijau)

kernel = np.ones((5, 5), np.uint8)
mask_bersih = cv2.morphologyEx(mask_awal, cv2.MORPH_CLOSE, kernel)

segmentasi_hijau = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_bersih)
```

```python
# Fathur Rasyid Irimawa 202431036
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

ax[0].imshow(img)
ax[0].set_title('Citra Asli')

ax[1].imshow(segmentasi_hijau)
ax[1].set_title('Segmentasi Daun')

plt.show()
```
